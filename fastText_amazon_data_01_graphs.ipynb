```python
def plot_df(df, title):
    xs = [i for i, _ in enumerate(df.n_rows)]
    plt.figure(figsize=[10,7])
    
    plt.plot(xs, df.acc, 'g-', label='Accuracy')
    plt.plot(xs, df.f1, 'r-', label='F1 score')
    
    for xy in zip(xs, df.acc):                                       
        plt.annotate('%s' % round(xy[1], 2), xy=xy)
    for xy in zip(xs, df.f1):                                       
        plt.annotate('%s' % round(xy[1], 2), xy=xy)
    
    plt.xlabel('the number of train samples')
    plt.xticks(xs, df.n_rows)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_df(movies_reviews_val, 'amazon movies reviews as is: validation results')
```
