```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['100k', 0.928128, 0.959207],
                 ['200k', 0.930059, 0.960317],
                 ['500k', 0.933335, 0.962104],
                 ['1M', 0.934667, 0.962847],
                 ['all', 0.935113, 0.963085]], columns = ['n_rows', 'acc', 'f1'])
b = DataFrame([['100k', 0.641839, 0.723253],
                 ['200k', 0.662383, 0.733861],
                 ['500k', 0.682364, 0.74526],
                 ['1M', 0.692777, 0.751121],
                 ['all', 0.697655, 0.753725]], columns = ['n_rows', 'acc', 'f1'])
```

```python
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, a.n_rows)
plt.title("test on amazon data")
plt.legend()
plt.show()
```

```python
xs = [i for i, _ in enumerate(b.acc)]
plt.plot(xs, b.acc, 'g-', label='Accuracy')
plt.plot(xs, b.f1, 'r-', label='F1 score')
# plt.axis([0, 4, .9, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, b.n_rows)
plt.title("test on imdb data")
plt.legend()
plt.show()
```
